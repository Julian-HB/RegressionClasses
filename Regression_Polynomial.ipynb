```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# Gegebene Daten
X = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
Y = np.array([2.5, 5.3, 8.8, 16.3, 26.5, 40.2, 58.7])

# Polynomiale Features erstellen
degree = 2  # Grad des Polynoms
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

# Lineare Regression anwenden
model = LinearRegression()
model.fit(X_poly, Y)

# Koeffizienten der Regression erhalten
coefficients = model.coef_
intercept = model.intercept_

# Mathematische Funktion ausgeben
function = f"{intercept:.2f}"
for i in range(1, degree + 1):
    function += f" + {coefficients[i]:.2f} * x^{i}"

print(f"Polynomiale Funktion: y = {function}")

# Daten plotten
plt.scatter(X, Y, label='Datenpunkte')
x_range = np.linspace(0, 8, 100).reshape(-1, 1)
x_range_poly = poly_features.transform(x_range)
y_pred = model.predict(x_range_poly)
plt.plot(x_range, y_pred, color='red', label='Polynomiale Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Polynomiale Regression')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Daten erstellen
x_values = np.array([1, 2, 3, 4, 5, 6, 7])
y_values_actual = np.array([2, 7, 30, 52, 73, 82, 121])
y_values_regression = 2.04 + x_values ** 2.16

# DataFrame erstellen
df = pd.DataFrame({'x': x_values, 'Actual': y_values_actual, 'Regression': y_values_regression})

# Plot erstellen
plt.figure(figsize=(8, 8))
plt.plot(df['x'], df['Regression'], label='Regressionsgraph: $y = 2.04 + x^{2.16}$', color='black', linewidth=2)
plt.scatter(df['x'], df['Actual'], label='Tatsächliche Werte', color='blue', marker='o')

# Achsenbeschriftungen und Legende hinzufügen
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc='upper left')

# Plot anzeigen
plt.show()
```
