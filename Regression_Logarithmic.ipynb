```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import numpy as np

# Assuming you have loaded or created your data
x = np.array([1, 2, 3, 4])
y = np.array([0, 0, 1, 1])  # Assuming binary classification

X = x.reshape(-1, 1)
model = LogisticRegression(random_state=0).fit(X, y)

coefficients = model.coef_
intercept = model.intercept_

print("Das sind die Parameter: ", coefficients, intercept)

# Generate points along the X-axis for plotting the logistic regression curve
x_values = np.linspace(min(x), max(x), 100).reshape(-1, 1)
y_prob = model.predict_proba(x_values)[:, 1]  # Probability of class 1

plt.scatter(X, y, color='orange', label='Test')
# Plot the logistic regression curve
plt.plot(x_values, y_prob, color='blue', label='Logistic Regression Curve')

# Set labels and title
plt.xlabel('x')
plt.ylabel('Probability of class 1')
plt.title('Logistic Regression with Scikit-learn')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
