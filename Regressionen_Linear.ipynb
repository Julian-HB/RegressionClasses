```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Create a DataFrame with the dataset
data = {'X': [1, 2, 3, 4, 5, 6, 7],
        'Y': [2, 7, 30, 52, 73, 82, 121]}
df = pd.DataFrame(data)

# Define the power function (a*X^b) for the regression
def power_function(x, a, b):
    return a * np.power(x, b)

# Perform the power regression
params, covariance = curve_fit(power_function, df['X'], df['Y'])

# Extract the parameters (a and b) from the regression
a, b = params

# Create a scatter plot of the data
plt.scatter(df['X'], df['Y'], label='Data')

# Generate points for the regression line
x_range = np.linspace(min(df['X']), max(df['X']), 100)
y_pred = power_function(x_range, a, b)

# Plot the regression line
plt.plot(x_range, y_pred, 'r', label=f'Regression Line: {a:.2f} * X^{b:.2f}')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Show the plot
#plt.grid()
plt.show()
print(covariance)
```
